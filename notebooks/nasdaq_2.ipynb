```python
import yfinance as yf
import datetime
import os

start_date = "2020-01-01"
end_date = datetime.datetime.today().strftime('%Y-%m-%d')

ticker = yf.Ticker("^GSPC")  # ^GSPC traks S&P 500 while QQQ tracks NASDAQ-100
sp500 = ticker.history(start=start_date, end=end_date)

os.makedirs("data/raw", exist_ok=True) # Ensure the data/ folder exists 
```

```python
sp500.tail(10)  # Show first few rows
```

```python
sp500.to_csv("data/raw/sp500.csv", index=True)
```
